### Data Preprocessing

In this section, we perform several steps to clean and prepare the dataset for analysis and modeling. These steps include data transformation, removal of unnecessary columns, and filtering of data to ensure it's relevant.

Steps performed:
1. **Precipitation Conversion**: The `PRECTOT` (total precipitation) variable is converted from its original unit (in seconds) by multiplying it by the number of seconds in a month (30 days), making it more meaningful for analysis.
2. **Remove Unnecessary Columns**: The columns `day`, `hour`, and `week_day` are removed as they are not useful for the analysis.
3. **Filter Data for Brazil**: We filter the dataset to only include data within the geographical bounds of Brazil (latitude between -33 and 5, longitude between -74 and -35).
4. **Time Column Conversion**: The "time" column, which was previously in a string format, is converted to `datetime` to allow for time-based operations.
5. **Temperature Conversion**: The temperature variable `T2M` is converted from Kelvin to Celsius to make it more intuitive for interpretation.
6. **Humidity Conversion**: The humidity variable `QV2M` is converted from a decimal to a percentage.
7. **Create Date-related Variables**: We extract the year and month from the "time" column, as these are useful features for temporal analysis.
8. **Create Temperature Range**: A margin of error is added to the temperature, with a minimum temperature (`T2M_min`) calculated by subtracting 3Â°C, and a maximum temperature (`T2M_max`) calculated by adding 4Â°C to the original temperature.

Finally, we display the first few rows of the preprocessed data to verify the changes.

```python
# Definir o nÃºmero de segundos em um mÃªs (30 dias)
seconds_in_month = 30 * 24 * 60 * 60

# Multiplicar PRECTOT pelo nÃºmero de segundos em um mÃªs
data['PRECTOT'] = (data['PRECTOT'] * seconds_in_month).round(2)

# Excluir as colunas 'day', 'hour' e 'week_day' 
colunas_para_excluir = ['day', 'hour', 'week_day']
colunas_existentes = [col for col in colunas_para_excluir if col in data.columns]

# Remover as colunas existentes
data = data.drop(columns=colunas_existentes)

# Excluir dados de fora do Brasil
data = data[(data['lat'] >= -33) & (data['lat'] <= 5) &  # Filtrar latitudes dentro do Brasil
            (data['lon'] >= -74) & (data['lon'] <= -35)]  # Filtrar longitudes dentro do Brasil

# Converter a coluna de 'time' para o formato datetime
data['time'] = pd.to_datetime(data['time'])

# Converter T2M de Kelvin para Celsius
data['T2M_Celsius'] = (data['T2M'] - 273.15).round(2)

# QV2M em %
data['QV2M_percentage'] = (data['QV2M'] * 100).round(3)

# Criar novas variÃ¡veis a partir da data
data['year'] = data['time'].dt.year
data['month'] = data['time'].dt.month

# Adicionar margem de erro de temperatura
data['T2M_min'] = (data['T2M_Celsius'] - 3).round(2)
data['T2M_max'] = (data['T2M_Celsius'] + 4).round(2)

# ===== CÃLCULO DA UMIDADE RELATIVA (UR%) =====
# FÃ³rmulas baseadas em QV2M (umidade especÃ­fica), PS (pressÃ£o) e T2M_Celsius (temperatura)

# PressÃ£o de vapor real (e)
data['e'] = (data['QV2M'] * data['PS']) / (0.622 + 0.378 * data['QV2M'])

# PressÃ£o de saturaÃ§Ã£o (es)
data['es'] = 610.78 * np.exp((17.27 * data['T2M_Celsius']) / (data['T2M_Celsius'] + 237.3))

# Umidade relativa em porcentagem
data['UR_percent'] = (data['e'] / data['es'] * 100).clip(0, 100).round(2)

# Remover colunas auxiliares (opcional)
data = data.drop(columns=['e', 'es'])

# Visualizar os dados apÃ³s o prÃ©-processamento
data.head()
```
